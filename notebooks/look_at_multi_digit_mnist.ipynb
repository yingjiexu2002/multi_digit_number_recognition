```python
import random

%matplotlib inline
import matplotlib.pyplot as plt

from handwritten_multi_digit_number_recognition.data import MultiDigitMNIST
```

```python
multi_digit_mnist = MultiDigitMNIST()
multi_digit_mnist.prepare_data()
multi_digit_mnist.setup()
```

```python
dataset = multi_digit_mnist.dataset["test"]
fig = plt.figure(figsize=(9, 9))
for i in range(6):
    ax = fig.add_subplot(6, 2, i + 1)
    rand_idx = random.randint(0, len(dataset) - 1)
    image, _ = dataset[rand_idx]
    ax.imshow(image.squeeze(), cmap="gray")
```
